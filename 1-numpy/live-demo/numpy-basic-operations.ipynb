# Mang con

Cu phap: 

```python 
x[start:stop:step]
x[:stop]
x[start:stop]
x[:stop:step]
x[::step]
x[start::step]
```

Mac dinh 
- `start = 0`
- `stop = {so phan tu}`
- `step = 1`

````python
arr = np.arange(1,11)
arr
````

````python
arr[2:7:2]
````

````python
arr[:5]
````

````python
arr[2:5]
````

````python
arr[::3]
````

````python
arr[2::4]
````

````python
arr[::]
````

# Noi mang

````python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
````

````python
concat = np.concatenate([x,y])
print(concat)
print(concat.shape)
````

````python
vstack = np.vstack([x,y])
print(vstack)
print(vstack.shape)
````

````python
hstack = np.hstack([x,y])
print(hstack)
print(hstack.shape)
````

````python
a = np.zeros((2,3))
b = np.ones((2,3))
````

````python
concat = np.concatenate((a,b))
print(concat)
print(concat.shape)
````

````python
concat = np.concatenate((a,b), axis=1)
print(concat)
print(concat.shape)
````

# Tach mang

````python
arr = np.arange(1,11)
print(arr)
````

````python
x1, x2, x3 = np.split(arr, [3,5])
print(x1)
print(x2)
print(x3)
````

````python
arr = np.random.randint(0, 10, size=(3, 4))
print(arr)
print(arr.shape)
````

````python
upper, lower = np.vsplit(arr, [2])
print(f"upper:\n{upper}")
print(upper.shape)
print(f"lower:\n{lower}")
print(lower.shape)
````

# Tinh toan

````python
a = np.array([1,2,3])
b = np.array([4,5,6])
````

## Tinh toan tren tung thanh phan

````python
r = a*b
print(r)
print(r.shape)
````

````python
r = np.multiply(a,b)
print(r)
print(r.shape)
````

## Tich vo huong

````python
r = np.matmul(a,b)
print(r)
print(r.shape)
````

````python
r = np.dot(a,b)
print(r)
print(r.shape)
````

## Tich co huong

````python
r = np.cross(a,b)
print(r)
print(r.shape)
````
